```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv("filepath")
data
```

```python
# Convert 'date' column to proper datetime format
data['date'] = pd.to_datetime(data['date'])

print(data.info())

data['formatted_date'] = data['date'].dt.strftime('%m/%d/%Y')

print(data[['date', 'formatted_date']])
```

```python
# Drop the 'formatted_date' column if not needed
data = data.drop(columns=['formatted_date'])

data['date'] = data['date'].dt.strftime('%m/%d/%Y')

# Check the data to confirm the index
data
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
filtered_data = (data['date'] >= pd.Timestamp("2024-05-15")) & (data['date'] <= pd.Timestamp("2024-12-31"))
```

```python
filtered_data = data.loc[filtered_data].copy()
```

```python
filtered_data
```

```python
df = pd.DataFrame(filtered_data)
#Convert 'date' to datetime format if not 
df['date'] = pd.to_datetime(df['date'])

#Melt the DataFrame to long format
long_df = pd.melt(df, id_vars=['date'], 
                  value_vars=[col for col in df.columns if col != 'date'], 
                  var_name='unique_id', 
                  value_name='y')

#Rename the date column to 'ds'
long_df.rename(columns={'date': 'ds'}, inplace=True)

# Display the resulting long format DataFrame
print(long_df) 
```

```python
targets = ['NP205_stage', 'P33_stage', 'G620_water_level', 'NESRS1', 'NESRS2']
filtered_targets = long_df[long_df['unique_id'].isin(targets)].copy()
```

```python
filtered_targets
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

filtered_targets['ds'] = pd.to_datetime(filtered_targets['ds'])

stations = ['NP205_stage', 'P33_stage', 'G620_water_level', 'NESRS1', 'NESRS2']

fig, axes = plt.subplots(len(stations), 1, figsize=(7, 8), sharex=True)

for ax, station in zip(axes, stations):
    df_s = filtered_targets[filtered_targets['unique_id'] == station].sort_values('ds')
    if df_s.empty:
        ax.text(0.5, 0.5, f'No data for {station}', ha='center', va='center')
        ax.set_title(station)
        continue

    ax.plot(df_s['ds'], df_s['y'], label='Observed', linewidth=1.5)
    ax.set_title(station, fontsize=12)
    ax.set_ylabel('Water Levels', fontsize=10)
    ax.grid(alpha=0.3)

# shared x formatting
axes[-1].set_xlabel('Date', fontsize=12)
locator = mdates.AutoDateLocator()
formatter = mdates.ConciseDateFormatter(locator)
axes[-1].xaxis.set_major_locator(locator)
axes[-1].xaxis.set_major_formatter(formatter)

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
fig, axes = plt.subplots(len(stations), 1, figsize=(10, 4 * len(stations)), sharex=True)

def prepare_df(path, model_label):
    df = pd.read_csv(path).copy()

    if 'h' in df.columns:
        df = df.rename(columns={'h': 'horizon'})
    elif 'horizon_step' in df.columns:
        df = df.rename(columns={'horizon_step': 'horizon'})

    if 'ds' in df.columns:
        df['ds'] = pd.to_datetime(df['ds'])
    elif 'day_index' in df.columns:
        df = df.rename(columns={'day_index': 'ds'})
        df['ds'] = pd.to_datetime(df['ds'])

    if 'horizon' in df.columns:
        df = df[df['horizon'] == 7]

    if 'prediction' in df.columns:
        df = df.rename(columns={'prediction': 'pred'})

    df['_model'] = model_label
    return df[['unique_id', 'ds', 'horizon', 'pred', 'truth', '_model']]

df_Bolt7 = prepare_df(
    "filepath", #csv file from chronos bolt baseline model results
    "Chronos"  # baseline
)
df_MIBOLT7 = prepare_df(
    "filepath",
    "MI-RAF"
)
df_SIMBOLT7 = prepare_df(
    "filepath",
    "SIM-RAF"
)
h_step=7

filtered_targets = filtered_targets.copy()
filtered_targets['ds'] = pd.to_datetime(filtered_targets['ds'])

# stations of interest
stations = ['NP205_stage', 'P33_stage', 'G620_water_level', 'NESRS1', 'NESRS2']

palette = {
    "Truth": "#262D7C",
    "Chronos": "#8E1D73FF",
    "MI-RAF":  "#379F43FF",
    "SIM-RAF": "#e88515ff"
}

for ax, station in zip(axes, stations):

    # Truth for the station
    df_truth = (
        filtered_targets[filtered_targets['unique_id'] == station]
        .sort_values('ds')
    )

    if df_truth.empty:
        ax.text(0.5, 0.5, f"No truth data for {station}", ha='center', va='center')
        ax.set_title(station.replace('_stage', '').replace('_', ' '))
        continue

    h_truth, = ax.plot(
        df_truth['ds'], df_truth['y'],
        label="Truth",
        color=palette["Truth"],
        linewidth=2.7,
        linestyle='-'
    )

    model_handles = []
    model_labels = []


    for model_df in [df_Bolt7, df_MIBOLT7, df_SIMBOLT7]:
        model_name = model_df['_model'].iloc[0]
        df_pred = model_df[model_df['unique_id'] == station].sort_values('ds')
        if df_pred.empty:
            continue

        last_pred_date = df_pred['ds'].max()
        df_truth_clipped = df_truth[df_truth['ds'] <= last_pred_date]
        merged = pd.merge(
            df_pred[['ds', 'pred']],
            df_truth_clipped[['ds', 'y']],
            on='ds',
            how='inner'
        )

        if merged.empty:
            h_model, = ax.plot(
                df_pred['ds'], df_pred['pred'],
                label=f"{model_name} (no overlap)",
                color=palette[model_name],
                linewidth=2.3
            )
            model_handles.append(h_model)
            model_labels.append(f"{model_name} no overlap")
            continue

        mae_val = mae(merged['y'], merged['pred'])
        rmse_val = np.sqrt(mse(merged['y'], merged['pred']))

        h_model, = ax.plot(
            df_pred['ds'],
            df_pred['pred'],
            label=model_name,
            color=palette[model_name],
            linewidth=2.7
        )
        model_handles.append(h_model)
        model_labels.append(f"{model_name}  (MAE={mae_val:.3f}, RMSE={rmse_val:.3f})")

    # X limits
    all_pred_dates = pd.concat([
        df_Bolt7[df_Bolt7['unique_id'] == station]['ds'],
        df_MIBOLT7[df_MIBOLT7['unique_id'] == station]['ds'],
        df_SIMBOLT7[df_SIMBOLT7['unique_id'] == station]['ds']
    ])
    if not all_pred_dates.empty:
        xmin = all_pred_dates.min() - pd.Timedelta(days=1)
        xmax = all_pred_dates.max() + pd.Timedelta(days=1)
        ax.set_xlim(xmin, xmax)

    # Y limits
    value_list = []
    if not df_truth.empty:
        value_list.append(df_truth['y'].to_numpy())
    for model_df in [df_Bolt7, df_MIBOLT7, df_SIMBOLT7]:
        tmp = model_df[model_df['unique_id'] == station]
        if not tmp.empty:
            value_list.append(tmp['pred'].to_numpy())
    if value_list:
        combined_vals = np.concatenate(value_list)
        yrange = combined_vals.max() - combined_vals.min()
        ax.set_ylim(combined_vals.min() - 0.05 * yrange, combined_vals.max() + 0.05 * yrange)

    ax.set_title(f"{station.replace('_stage', '').replace('_', ' ')}: Lead Time = {h_step}", fontsize=19, pad=10)
    ax.set_ylabel("Water Level", fontsize=22)
    ax.grid(alpha=0.25)

    # Legends
    leg1 = ax.legend(handles=[h_truth], labels=["Truth"], loc="upper left",
                     frameon=False, fontsize=16.5)

    if model_handles:
        leg2 = ax.legend(
            model_handles, model_labels,
            loc="lower right",
            frameon=True,
            fontsize=15.5,
            borderpad=0.3,
            labelspacing=0.6,
            handlelength=1.2,
            handletextpad=0.6,
        )
        for txt in leg2.get_texts():
            txt.set_ha("right")
        leg2._legend_box.align = "right"
        ax.add_artist(leg1)
    ax.tick_params(axis='x', rotation=45, labelsize=26)
    ax.tick_params(axis='y', labelsize=26)
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%d %b %Y"))
    plt.setp(ax.get_xticklabels(), fontsize=12, rotation=45, ha="right")
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.1f}'))
    ax.tick_params(axis='both', which='major', labelsize=18.5)
    fig.tight_layout()
plt.show()

```
