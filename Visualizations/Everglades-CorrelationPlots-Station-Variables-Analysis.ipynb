```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("filepath")
data
```

```python
# Convert 'date' column to proper datetime format
data['date'] = pd.to_datetime(data['date'])

# Verify the column types to confirm the 'date' is now datetime
print(data.info())

# When displaying or exporting, you can format the date as MM/DD/YYYY
data['formatted_date'] = data['date'].dt.strftime('%m/%d/%Y')

# Print the first few rows to see the formatted date
print(data[['date', 'formatted_date']])
```

```python
# Drop the 'formatted_date' column if not needed
data = data.drop(columns=['formatted_date'])

data['date'] = data['date'].dt.strftime('%m/%d/%Y')

# Check the data to confirm the index
data
```

```python
import seaborn as sns
```

```python
# Plot correlation heatmap
# Exclude the 'date' column and calculate the correlation matrix
corr_matrix = data.select_dtypes(include=['number']).corr()

plt.figure(figsize=(16, 14))  # Increase figure size
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=.5, annot_kws={"size": 8})  
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.yticks(rotation=0)  # Keep y-axis labels straight
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display

# Rename columns
data.rename(columns={'NP205_stage': 'NP205', 'P33_stage': 'P33', 'G620_water_level': 'G620',
                     'S12A_T': 'S12A', 'S12B_T': 'S12B', 'S12C_T': 'S12C'}, inplace=True)

# Define selected columns for both heatmaps
selected_columns1 = ['NP205', 'P33', 'G620', 'NESRS1', 'NESRS2', 'S12A', 'S12B', 'S12C', 'S12D_T', 'S333_T']
selected_columns2 = ['NP205', 'P33', 'G620', 'NESRS1', 'NESRS2']

# Compute correlation matrices
corr_matrix1 = data[selected_columns1].corr()
corr_matrix2 = data[selected_columns2].corr()

# Create first figure
fig1 = plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix1, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5,
            annot_kws={"size": 13.5})
plt.xticks(rotation=90, fontsize=13.5)
plt.yticks(rotation=0, fontsize=13.5)
plt.show()

# Create second figure
fig2 = plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix2, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5,
            annot_kws={"size": 13.5})
plt.xticks(rotation=90, fontsize=13.5)
plt.yticks(rotation=0, fontsize=13.5)
plt.show()
```
