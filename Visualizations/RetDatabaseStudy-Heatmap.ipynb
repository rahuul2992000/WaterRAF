```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# MAE values
mae_matrix = np.array([
    [0.072, 0.071, 0.071, 0.072, 0.074],  # Lead time = 7
    [0.107, 0.105, 0.108, 0.111, 0.116],  # Lead time = 14
    [0.142, 0.139, 0.146, 0.147, 0.160],  # Lead time = 21
    [0.173, 0.170, 0.175, 0.181, 0.200]   # Lead time = 28
])


x_labels = ['85%', '65%', '45%', '25%', '0%']
y_labels = ['7', '14', '21', '28']

fig, ax = plt.subplots(figsize=(32, 5))
im = ax.imshow(mae_matrix, cmap='YlGnBu_r', origin='lower')

# Annotate values with color based on row
for i in range(mae_matrix.shape[0]):
    for j in range(mae_matrix.shape[1]):
        value = mae_matrix[i, j]
        if i == 0:
            color = 'white'
        elif i in [1, 2]:
            color = 'black'
        else:
            color = 'Red'
        ax.text(j, i, f"{value:.3f}", ha='center', va='center', color=color, fontsize=18.2,  fontweight='bold')

# Set axis ticks and labels
ax.set_xticks(np.arange(len(x_labels)))
ax.set_xticklabels(x_labels, fontsize=20)
ax.set_yticks(np.arange(len(y_labels)))
ax.set_yticklabels(y_labels, fontsize=20)

# Labels with padding
ax.set_xlabel("Retrieval Pool Size", fontsize=23, labelpad=15)
ax.set_ylabel("Lead Time (days)", fontsize=23, labelpad=15)

# Colorbar with padding
cbar = plt.colorbar(im, ax=ax, pad=0.02)
cbar.set_label("MAE", fontsize=23, labelpad=15)
cbar.ax.tick_params(labelsize=15)
ax.set_aspect(0.7) 

# Adjust layout
plt.subplots_adjust(left=0.15, right=0.95, top=0.95, bottom=0.12)
plt.show()
```
